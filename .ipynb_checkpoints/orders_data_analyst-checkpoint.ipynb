```python
import zipfile
zip_ref = zipfile.ZipFile('orders.csv.zip')
zip_ref.extractall()
zip_ref.close()
```

```python
import pandas as pd
df = pd.read_csv('orders.csv', na_values=['Not Available', 'unknown'])
df['Ship Mode'].unique()

```

```python
df
```

```python
#df.rename(columns={'Order Id':'order_id', 'City':'city'})
df.columns=df.columns.str.lower()
df.columns=df.columns.str.replace(' ','_')
df.head(5)
```

```python
df['discount'] = df['list_price']*df['discount_percent']*.01
df['sale_price'] = df['list_price']-df['discount']
df['profit'] = df['sale_price']-df['cost_price']
df
```

```python
df['order_date'] = pd.to_datetime(df['order_date'], format="%Y-%m-%d")
df
```

```python
df.drop(columns=['list_price','cost_price','discount_percent'], inplace=True)
df
```

```python
import sqlalchemy as sal

# Create the engine
engine = sal.create_engine('postgresql+psycopg2://f3i:admin@localhost:5432/orders_db')

# Connect to the database
conn = engine.connect()

# Upload the DataFrame to a new table called 'my_table' in the new_db database
df.to_sql('my_table', con=conn, index=False, if_exists='replace')

# Close the connection
conn.close()
```
